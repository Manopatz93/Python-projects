```python
df.drop(['Row ID','Order ID','Customer ID','Customer Name'],axis=1,inplace=True)
```

```python
df.drop_duplicates(inplace=True)
```

```python
df["Postal Code"].fillna(5401,inplace=True)
```

```python
df.nunique()
```

```python
# first we need to change the data type of the order and ship date into date type
df[["Order Date","Ship Date"]].head()
```

```python
df['Order Date']=pd.to_datetime(df['Order Date'])
df['Ship Date']=pd.to_datetime(df['Ship Date'])
```

```python
df.dtypes
```

```python
# data cleaning process is over , as we can see our data is good to do the next process
```

```python
top=df.groupby("City").sum().sort_values("Sales",ascending=False).head(15).round(0)
```

```python
top.head()
```

```python
top=top.reset_index()
```

```python
import plotly.express as px
```

```python
fig=px.bar(top,x='City',y='Sales',title="Top 15 Cities",color_discrete_sequence=px.colors.sequential.RdBu)
fig.show()
```

```python
# then we will do the same for state
```

```python
top_state=df.groupby("State").sum().sort_values("Sales",ascending=False).head(15).round(0)
top_state.reset_index(inplace=True)
top_state
```

```python
fig1=px.bar(top_state,x="State",y="Sales",title="top state reg sales")
fig1.show()
```

```python
# now we gonnna do the category based on sales
```

```python
top_cat=df.groupby("Category").sum().sort_values("Sales").round(0)
top_cat.reset_index(inplace=True)
```

```python
# sub category
fig6= px.pie(df, values=np.ones(9799), names='Sub-Category', title= "Subcategory")
fig6.show()
```

```python
# SALES BASED ON REGION 
```

```python
top_region= df.groupby("Region").sum().sort_values("Sales",ascending=False).round(0)
top_region.reset_index(inplace=True)
fig6=px.pie(top_region, values="Sales",names= "Region",title="Sales based on region")
fig6.show()
```

```python
# SALES BASED ON SEGMENTS
```

```python
top_sale_seg= df.groupby("Segment").sum().sort_values("Sales",ascending=False).round(0)
top_sale_seg.reset_index(inplace=True)
fig8= px.pie(top_sale_seg, values= "Sales", names= "Segment", title= " sales on each segments")
fig8.show()
```

```python
# top cities based on consumer and with reference of category
```

```python
plt.figure(figsize=(18,10))
city_cat=df.groupby(["City","Category"]).sum().sort_values("Sales",ascending=False).head(30).round(0)
city_cat.reset_index(inplace=True)
sns.barplot(data=city_cat,x="City",y="Sales",hue="Category")
plt.xticks(rotation='vertical')
plt.show()
```

```python
plt.figure(figsize=(18,10))
city_seg=df.groupby(["City","Segment"]).sum().sort_values("Sales",ascending=False).head(30).round(0)
city_seg.reset_index(inplace=True)
sns.barplot(data=city_seg,x="City",y="Sales",hue="Segment")
plt.xticks(rotation='vertical')
plt.show()
```
